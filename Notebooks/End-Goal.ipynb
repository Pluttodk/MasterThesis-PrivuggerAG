```python
import attacker
from typing import *
from sklearn.feature_selection import mutual_info_regression
import numpy as np
```

```python
def f(age : List[float]) -> float:
    return sum(age)/len(age)
```

```python
domain = [
    {
        "name": "age",
        "lower": 10,
        "upper": 80,
        "type": "float"
    },
]
```

```python
def q(trace):
    I = mutual_info_regression(trace["Alice_age"].reshape(-1,1), trace["out"], discrete_features=False)[0]
    return -I
```

```python
res = attacker.construct_analysis(f, domain, q)
```

```python
res.run()
```

```python
res.plot_convergence()
```
