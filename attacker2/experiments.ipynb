```python
plt.rcParams["figure.figsize"] = (20,8)
for rv in model.free_RVs:
    plt.plot(x, np.exp(rv.distribution.logp_nojac(x).eval()), label=rv.__str__())
plt.legend(loc=2)
plt.tight_layout()
```
